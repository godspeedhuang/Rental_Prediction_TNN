```python
from collections import Counter
# look for outliers
def detect_outliers(data , n ,features):
    outliers_indices = []

    for col in features:
        Q1 = np.percentile(data[col],25)
        Q3 = np.percentile(data[col],75)
        IQR = Q3 - Q1
        outlier_step = 1.5*IQR #離群值為1.5倍四分位距
        outlier_list_col = data[(data[col] < Q1 - outlier_step) | (data[col] > Q3 + outlier_step)].index
        outliers_indices.extend(outlier_list_col)

    #尋找超過n次的資料(n+1個特徵為outlier)
    outliers_indices = Counter(outliers_indices)
    multiple_outlier = list(k for k,v in outliers_indices.items() if v>n)

    return multiple_outlier    
```

```python
needdellist = detect_outliers(rentdata,0,['每坪租金（含管理費與停車費）','坪數'])
print(needdellist)
print(f'總共 {len(needdellist)} 個離群值')
```

```python
from sklearn.model_selection import train_test_split
x = rentdata.drop('每坪租金（含管理費與停車費）',axis = 1)
y = rentdata['每坪租金（含管理費與停車費）']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=10)
```

```python
import statsmodels.api as sm
x_train = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train)
results = model.fit()
results.summary()
```

```python
x_test = sm.add_constant(x_test)
statmodels_pred = results.predict(x_test)
from sklearn import metrics
#MAE(平均誤差)
print('MAE =' , metrics.mean_absolute_error(y_test,statmodels_pred))
#MSE(均方誤差)
print('MSE =' , metrics.mean_squared_error(y_test,statmodels_pred))
#RMSE(平方根誤差)
print('RMSE = ' , np.sqrt(metrics.mean_squared_error(y_test,statmodels_pred)))
#MAPE
print('MAPE = ' ,round((np.mean(np.abs((y_test-statmodels_pred)/y_test))*100),3),'%')
```

```python
plt.scatter(y_test,statmodels_pred,c = 'b')
plt.xlabel('測試集實際值',FontProperties = font)
plt.ylabel('線性迴歸預測值',FontProperties = font)
plt.show
```

```python
#lasso懲罰項為1

from sklearn.linear_model import Lasso , LassoCV
lasso_model = Lasso().fit(x_train,y_train) #默認ALPHA=1
lasso_pred = lasso_model.predict(x_test)
coeff_used = np.sum(lasso_model.coef_!= 0)
print(np.sqrt(metrics.mean_squared_error(y_test,lasso_pred))) #RSME
print(metrics.r2_score(y_test, lasso_pred)) #R2
print(coeff_used) #變數使用
pd.set_option('display.max_rows',None)
print(pd.Series(lasso_model.coef_, index = x_train.columns)) #變數顯示
print('MAPE = ' ,round((np.mean(np.abs((y_test-lasso_pred)/y_test))*100),3),'%') #MAPE
```

```python
def test_Lasso_alpha(*data):
    X_train,X_test,y_train,y_test=data
    alphas = [0.01,0.02,0.05,0.1,0.2,0.5,1,2,5,10,20,50,100,200,500,1000]
    scores = []
    for i,alpha in enumerate(alphas):
        regr = Lasso(alpha=alpha)
        regr.fit(X_train,y_train)
        scores.append(regr.score(X_test,y_test))
    ##繪圖
    plt.rcParams['axes.unicode_minus'] = False
    fig = plt.figure()
    ax=fig.add_subplot(1,1,1)
    ax.plot(alphas,scores)
    ax.set_xlabel(r"$\alpha$")
    ax.set_ylabel(r"score")
    ax.set_xscale('log')
    ax.set_title("Lasso")
    plt.show()
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=10)
test_Lasso_alpha(X_train,X_test,y_train,y_test)
```
