```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('前處理後資料_統整.csv')
data
```

```python
data_district=data.groupby(by=['鄉鎮市區']).agg({'每坪租金（含管理費與停車費）':np.mean})
data_district
```

```python
data_district['區']=data_district.index.str[3:]
```

```python
data_district
data_district.to_csv('區平均租金.csv')
```

```python
data_vill=data.groupby(by=['VILLNAME']).agg({'每坪租金（含管理費與停車費）':np.mean})
data_vill
```

```python
data_vill['里']=data_vill.index
data_vill
```

```python
data_vill.to_csv('里平均租金.csv')
```
