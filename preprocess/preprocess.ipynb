```python
import pandas as pd
import numpy as np
```

```python
# 以2021年01-05月份資料為主
data_Q1=pd.read_csv(r'raw_data\2021Q1-deduplicated.csv')
data_M4=pd.read_csv(r'raw_data\202104-deduplicated.csv')
data_M5=pd.read_csv(r'raw_data\202105-deduplicated.csv')
```

```python
# CONCAT 3 FILES
data_M1_M5=pd.concat([data_Q1,data_M4,data_M5],axis=0)
data_M1_M5
```

```python
# SELECT COUNTY
data_M1_M5=data_M1_M5[data_M1_M5['縣市']==18]
data_M1_M5
```

```python
# GET TARGET TOWN CODE
town_list=['臺南市東區','臺南市中西區','臺南市北區','臺南市南區','臺南市安平區','臺南市安南區','臺南市永康區']
towncode_list=pd.read_csv(r'編碼表\編碼表.鄉鎮市區.csv')
towncode_list=towncode_list[towncode_list['代表'].isin(town_list)]
towncode_list=towncode_list['編碼'].values
towncode_list
```

```python
# SELECT TOWN
data_M1_M5=data_M1_M5[data_M1_M5['鄉鎮市區'].isin(towncode_list)]
data_M1_M5
```

```python
# GET NA
data_M1_M5=data_M1_M5.replace('-',None)
```

```python
# GET NA NUMBER IN EACH COLUMN
na_values=data_M1_M5.isna().sum().values
na_values
```

```python
# SELECT COLUMNS WHICH NA NUM > 9000
na_9000=data_M1_M5.iloc[:,np.argwhere(na_values>9000).T[0]]
na_9000
```

```python
# DROP THESE COLUMNS
data_M1_M5=data_M1_M5.drop(na_9000,axis=1)
```

```python
import re
pattern=r'\((\d+.\d+), (\d+.\d+), (\d+.\d+), (\d+.\d+)\)'
lat_list=[]
lng_list=[]
for i in data_M1_M5['約略地點範圍']:
    t=re.search(pattern,i)
    lat=t.group(1)
    lng=t.group(2)
    lat_list.append(lat)
    lng_list.append(lng)
```

```python
data_M1_M5['緯度']=lat_list
data_M1_M5['經度']=lng_list
```

```python
def get_code(name):
    data=pd.read_csv(f'編碼表\\編碼表.{name}.csv')
    data=data[['編碼','代表']].to_dict('tight')['data']
    list1=[i[0] for i in data]
    list2=[i[1] for i in data]
    data=dict(zip(list1,list2))
    return data
```

```python
type_dic=get_code('性別限制')
data_M1_M5['性別限制']=data_M1_M5['性別限制'].apply(lambda x:type_dic[x])
data_M1_M5['性別限制']
```
